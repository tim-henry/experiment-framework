```python
n = 10
a = np.zeros((n, n))

dataset = 3

for k in range(dataset):
    x = np.floor(k / 2).astype(int) + 1 if k % 2 == 1 else 0
    y = np.floor(k / 2).astype(int)     if k % 2 == 0 else 0
    print(x, y)
    for i in range(n):
        a[(x + i) % n, (y + i) % n] = 1
        
plt.imshow(a)
        
```
