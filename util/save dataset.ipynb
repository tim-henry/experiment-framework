```python
import os
os.chdir("/Users/timhenry/Documents/mit/meng/src")
import numpy as np
import torch
import config.input
```

```python
_, test_loader_fn = config.input.options["col_loc_scale_mnist"]
args = {
    "batch_size": 1,
    "test_batch_size": 256,
    "use_cuda": False,
    "keep_pct": 9, 
    "color_indices": np.arange(9),
    "example_pct": 1,
}
test_loader = test_loader_fn(args)
```

```python
out = torch.zeros((0, 3, 32, 32))
out_t = torch.zeros((0, 3)).long()
for data, target in test_loader:
    out = torch.cat((out, data))
    out_t = torch.cat((out_t, target))
```

```python
import pickle
with open("data.p", 'wb') as file:
    pickle.dump(out, file)
    
with open("targets.p", 'wb') as file:
    pickle.dump(out_t, file)
```

```python
data = pickle.load(open("data.p","rb"))
targets = pickle.load(open("targets.p","rb"))

import torch.utils.data as utils
loader = utils.DataLoader(utils.TensorDataset(data, targets), batch_size=1)
```
